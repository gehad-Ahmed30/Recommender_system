## Building Our Course Recommender (Cosine Similarity)

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cos_sim_m = cosine_similarity(cv_m)
cos_sim_m
```

```python
scores = list(enumerate(cos_sim_m[index]))
scores
```

```python
sorted_score = sorted(scores, key=lambda x: x[1], reverse=True)
sorted_score
```

```python
sorted_indices = [i[0] for i in sorted_score[1:]]
sorted_indices
```

```python
recommended_result_df = data.iloc[sorted_indices]
recommended_result_df.head()
```

```python
recommended_result_df['Similarity_Score'] = np.array(sorted_score[1:])[:,1]
recommended_result_df.head()
```

```python
use_df = recommended_result_df[['Clean_title','Similarity_Score']]
use_df
```

##### Putting it all Together

```python
def recommend_course(title,numrec=10):
    course_index = pd.Series(
        data.index, index=data['course_title']).drop_duplicates()

    index = course_index[title]

    scores = list(enumerate(cos_sim_m[index]))

    sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)

    selected_course_index = [i[0] for i in sorted_scores[1:]]

    selected_course_score = [i[1] for i in sorted_scores[1:]]

    rec_df = data.iloc[selected_course_index]

    rec_df['Similarity_Score'] = selected_course_score

    final_recommended_courses = rec_df[[
        'course_title', 'Similarity_Score', 'url', 'price', 'num_subscribers']]

    return final_recommended_courses.head(numrec)
```

```python
result = recommend_course('Beginner to Pro - Financial Analysis in Excel 2017',3)
result
```
